### 01) WAP to demonstrate the use of Pie chart.

```python
x=[2000,3000,3000,2000]
y=["Food","Travel","Accomodation","Shopping"]
c=["r","y","g","b"]
ex=[0.2,0,0,0]
plt.pie(x,labels=y,colors=c,explode=ex)
plt.legend(y,title ="Travel Expenses",loc ="center left",bbox_to_anchor =(1, 0, 0.5, 1))
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=["Kohli","Rohit","Dhoni","Gill"]
y=[3500,1500,1000,2000]
c=["r","g","b","c"]
plt.bar(x,y,color=c,alpha=1,align="edge")
plt.xlabel("Crickter Names")
plt.ylabel("No. RUns")
plt.title("Crickter Runs In One Year")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

x=["Kohli","Rohit","Dhoni","Gill"]
y=[3500,1500,1000,2000]
c=["r","g","b","c"]
plt.bar(x,y,color=c,alpha=1,align="edge")
plt.xlabel("Crickter Names")
plt.ylabel("No. RUns")
plt.title("Crickter Runs In One Year")
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=["Kohli","Rohit","Dhoni","Gill"]
y=[3500,1500,1000,2000]
size=[100,200,300,400]
plt.scatter(x,y,c="r",marker="*",s=size)
plt.xlabel("Crickter Names")
plt.ylabel("No. RUns")
plt.title("Crickter Runs In One Year")
plt.savefig("mychart.jpg")
```
